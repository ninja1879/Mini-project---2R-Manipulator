```python
import numpy as np
import matplotlib.pyplot as plt

# Manipulator parameters
l1 = 1.0  # Length of the first link
l2 = 0.7  # Length of the second link

# Joint angle limits in degrees
theta1_min = 35
theta1_max = 145
theta2_min = 35
theta2_max = 145

# Convert angles to radians
theta1_range = np.deg2rad(np.arange(theta1_min, theta1_max + 1))
theta2_range = np.deg2rad(np.arange(theta2_min, theta2_max + 1))

# Initialize lists to store workspace points
workspace_x = []
workspace_y = []

# Calculate workspace points
for theta1 in theta1_range:
    for theta2 in theta2_range:
        x = l1 * np.cos(theta1) + l2 * np.cos(theta1 + theta2)
        y = l1 * np.sin(theta1) + l2 * np.sin(theta1 + theta2)
        workspace_x.append(x)
        workspace_y.append(y)

# Plot the workspace
plt.figure(figsize=(8, 6))
plt.scatter(workspace_x, workspace_y, s=5, color='g', marker='.')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('Workspace of 2R Elbow Manipulator')
plt.grid(True)
plt.show()
```
